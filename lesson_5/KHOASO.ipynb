#**Abstraction**

 - Find the lock number that divisiable to 3 and have the largest string.

#**Input**

 - An integer number with length more than 2 digits.

#**Output**
  
 - A number that divisiable to 3 and have the largest string.

#**Pattern Recognition**

 - Dynamic Programming

#**Pseudocode**

        1. Enter and split an integer number as arr.
        2. Sort arr descendingly.
        3. Create 2 lists q1 and q2 to store digit that divide 3 by 1 and by 2 sequentially.
        4. Divide each element in arr with 3 to get the remainder and arrange them into q1 and q2.
        5. Remainder = sum of arr % 3
        6. if Remainder == 1:
        7. 		Check if q1 is empty:
        8.			Remove all element in q2.
        9.		if not remove the minimum element in q1.
        10.	elif Remainder == 2:
        11.		Check if q2 is empty:
        12.			Remove 2 minimum elements in q1
        13.		if not remove the minimum element in q2.

#**Source Code**

```
arr = list(map(int, input().strip()))
q1 = []
q2 = []
def isPossibleToMakeDivisible(arr):
    arr = sorted(arr, reverse=True)
    for i in arr:
        r = i % 3
        if r == 1:
            q1.append(i)
        elif r == 2:
            q2.append(i)
    r = sum(arr) % 3
    if r == 1:
        if not q1:
            for i in q2:
                arr.remove(i)
        else:
            arr.remove(min(q1))
    if r == 2:
        if not q2:
            arr.remove(min(q1))
            q1.remove(min(q1))
            arr.remove(min(q1))
        else:
            arr.remove(min(q2))
    return arr
arr = isPossibleToMakeDivisible(arr)
s = ""
for ele in arr:
    s += str(ele)
print(s)
```
