#**Abstraction**
Cho 5 số a, k, b, m, n. Tìm số nguyên $x$ nhỏ nhất sao cho:$$S(x) = a \times (x - \lfloor \frac{x}{k} \rfloor) + b \times (x - \lfloor \frac{x}{m} \rfloor) \geqslant n$$

#**Input**
Nhập vào 5 số nguyên dương a, k, b, m, n.
#**Output**
Một số nguyên là số ngày tính được.
  
#**Pattern Recognition**
Dynamic Programming

#**Pseudocode**

> Nhập vào 5 số nguyên dương a, k, b, m, n

> Gán left = 0

> Gán right = 10^18

> Gán num_day = -1

> while left <= right:

> > Gán mid = (left + right) //2

> > Gán sum = a*(mid - mid//k) + b*(mid - mid//m)

> > Kiểm tra nếu sum lớn hơn hoặc bằng n:

> > > Gán num_day = mid

> > > Gán right = mid - 1

> > Nếu sum nhỏ hơn n:

> > > left = mid + 1

> Xuất giá trị của num_day

#Code


```
(a, k, b, m, n) = list(map(int, input().split()))

left = 0
right= 10**18
num_day = -1

while left <= right:
    mid = (left + right) //2
    sum = a*(mid - mid//k) + b*(mid - mid//m)
    if sum >= n:
        num_day = mid
        right = mid - 1
    else:
        left = mid + 1
        
print(num_day)
```
