#**Abstraction**

  *  Có hai phân số a/b và c/d. Mối lần biến đổi là tăng giá trị a và b lên 1, giản ước phân số vừa nhận được. Tính số lần biến đổi để a/b = c/d.

#**Input**

 * Nhập bốn số tự nhiên a, b, c, d.

#**Output**
  
  *  Số lần cần phải biến đổi a và b.
  
#**Pattern Recognition**

 Dynamic Programming

#**Pseudocode**


> Nhập bốn số a, b, c, d.

> Kiểm tra a*d < b*c và c/d < 1 và a/b < 1

> Khai báo n = 0

> While a*d <= b*c:

> > Tăng giá trị của a lên 1

> > Tăng giá trị của b lên 1

> > Gán giá trị ước chung lớn nhất của a và b cho biến uoc

> > Chia làm tròn cho a và uoc và cập nhật lại a

> > Chia làm tròn cho b và uoc và cập nhật lại b

> > Tăng giá trị n lên 1

> > Kiểm tra nếu a == c và b == d:

> > > Xuất giá trị biến n

> > > Kết thúc chương trình

**Code**

```
a = int(input())
b = int(input())
c = int(input())
d = int(input())
def ucln(a, b):
    if (b == 0):
        return a
    return ucln(b, a % b)

if (a*d < b*c and c/d < 1 and a/b < 1):
  n = 0
  while (a*d <= b*c):
      a = a+1
      b = b+1
      uoc = ucln(a,b)
      a = (a)//uoc
      b = (b)//uoc
      n+=1
      if (a==c and b == d):
        print(n)
        quit()
```
