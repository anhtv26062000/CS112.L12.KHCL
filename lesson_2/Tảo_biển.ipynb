**Pattern Recognition**

Quy hoạch động

Ý tưởng:

Khởi tạo 2 biến tạm được cập nhật liên túc, 1 biến sẽ lưu tổng giá trị trước đó, một biến lưu tổng giá trị hiện tại.

**Mã giả :**

Input : hai số nguyên: n, k (1 ≤ n ≤ 1000, 1 ≤ k ≤ 1018)

Output : Một số nguyên.

Gán pre_sum = 0 , cur_sum=1.

Chạy vòng for từ 1 đến n:

      Gán k = cur_sum

      Gán pre_sum = = pre_sum + cur_sum
      
      Gán cur_sum = pre_sum + k

Xuất ra kết quả.



  



```
n = input()
n=list(map(int,n.split()))
pre_sum=0
cur_sum=1
for i in range(n[1]):
    k=cur_sum
    pre_sum = pre_sum + cur_sum
    cur_sum = pre_sum +k
print((cur_sum*n[0])%1000000007)
```
